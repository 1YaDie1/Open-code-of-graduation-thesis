# 机器学习分类模型

## 数据预处理：70%训练数据；30%测试数据

```python
BL = "实践风格" # 逻辑风格、感官风格、计划风格、实践风格
data = pd.read_excel("./test.xlsx") # 测试哪个风格就导入哪个风格数据集
data
```

```python
# 按行打乱所有数据
shuffled_data = data.sample(frac=1, random_state=42)  # random_state 用于设置随机种子 frac=1 表示对所有行进行重排

# 重置索引（可选）
shuffled_data.reset_index(drop=True, inplace=True)
shuffled_data
```

```python
X_data = shuffled_data.loc[:,"第1个行为":"第32个行为"].values
X_data.shape
```

```python
target = shuffled_data.loc[:,BL].values
target.shape
```

```python
# 切分数据
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X_data,target,test_size=0.3)

display(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

## KNN模型训练数据

```python
from sklearn.neighbors import KNeighborsClassifier # 导入KNN分类库
from sklearn.metrics import classification_report # 生成分类报告

# 1、引入模型
KNN = KNeighborsClassifier()
# 2、训练模型
model = KNN.fit(x_train,y_train)

# 3、计算测试集准确率（accuracy）
print(f'knn测试集准确率：{model.score(x_test,y_test)}')

# 4、打印测试集分类报告
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```

## 贝叶斯模型训练数据

```python
# 导入贝叶斯算法(伯努利模型)
from sklearn.naive_bayes import BernoulliNB

# 1、引入模型
BE = BernoulliNB()
# 2、训练模型
model = BE.fit(x_train,y_train)

# 3、计算测试集准确率（accuracy）
print(f'贝叶斯测试集准确率：{model.score(x_test,y_test)}')

# 4、打印测试集分类报告
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```

## 决策树模型训练数据

```python
from sklearn.tree import DecisionTreeRegressor

# 1、引入模型
DT = DecisionTreeRegressor()
# 2、训练模型
model = DT.fit(x_train,y_train)

# 3、计算测试集准确率（accuracy）
print(f'决策树测试集准确率：{model.score(x_test,y_test)}')

# 4、打印测试集分类报告
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```
